### 1. Load and Review the Dataset  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv(r'Raw Data\Employees_Attirition.csv')
```

---

### 2. Data Cleaning 

```python
# Check for missing values
df.isnull().sum()
```

```python
# Visualize missing data
sns.heatmap(df.isnull(), cbar=True, cmap="viridis")
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Handle missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['JobSatisfaction'].fillna(df['JobSatisfaction'].mode()[0], inplace=True)
df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean(), inplace=True)
df.dropna(inplace=True)
```

---

### 3. Feature Engineering  

```python
# Create tenure buckets
df['Tenure_Bucket'] = pd.cut(
    df['YearsAtCompany'],
    bins=[0, 3, 7, 15, 40],
    labels=['0-3 years', '4-7 years', '8-15 years', '16+ years']
)
```

```python
# Encode target variable
df['Attrition_Encoded'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Interaction features
df['WorkLifeBalance_JobSatisfaction'] = df['WorkLifeBalance'] * df['JobSatisfaction']
df['WorkLifeBalance_JobSatisfaction']
```

---

### 4. Univariate and Multivariate Analysis  


```python
# Univariate Analysis
sns.countplot(data=df, x='Attrition')
plt.title('Attrition Distribution')
plt.show()
```

```python
# Numerical feature distributions
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title('Monthly Income Distribution')
plt.show()
```

```python
# Pairwise relationships
features = [
    'Age', 'YearsAtCompany', 'TotalWorkingYears', 'MonthlyIncome',
    'YearsInCurrentRole', 'JobSatisfaction', 'Attrition']
sns.set_theme(style="whitegrid", palette="muted")
sns.pairplot(df[features], 
             hue='Attrition', 
             diag_kind='kde', 
             markers=["o", "s"], 
             plot_kws={'alpha': 0.7, 's': 50})
plt.suptitle("Pairplot of Employee Attrition", size=16, y=1.02)
plt.show()
```

```python
effective_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'YearsInCurrentRole']

correlation_matrix = df[effective_features].corr(numeric_only=True)

plt.figure(figsize=(8, 6))

sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap="coolwarm", 
            vmin=-1, vmax=1, 
            linewidths=0.5, 
            cbar_kws={'shrink': 0.8}, 
            annot_kws={"size": 14, "weight": "bold", "color": "white"})

plt.title("Correlation Heatmap", fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)

plt.tight_layout()
plt.show()
```

### 5. Insights Worth Mentioning

```python
# Grouping by department and attrition
dept_attrition = df.groupby(['Department', 'Attrition']).size().unstack()
dept_attrition
```

```python
# Pivot table for average income by department and attrition
pivot_table = df.pivot_table(
    values='MonthlyIncome',
    index='Department',
    columns='Attrition',
    aggfunc='mean'
)
pivot_table
```

```python
sns.heatmap(pivot_table, cmap="Blues")
plt.title("Average Monthly Income by Department and Attrition")
plt.show()
```

---

### 6. Outlier Detection and Handling  

```python
# Outlier detection using IQR (Interquartile Range)
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[
    (df['MonthlyIncome'] < Q1 - 1.5 * IQR) |
    (df['MonthlyIncome'] > Q3 + 1.5 * IQR)
]
outliers
```

```python
# Visualize outliers
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income by Attrition')
plt.show()
```

```python
# Remove outliers
data_cleaned = df[
    (df['MonthlyIncome'] >= Q1 - 1.5 * IQR) &
    (df['MonthlyIncome'] <= Q3 + 1.5 * IQR)
]
```
